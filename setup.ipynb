# Setup Notebook

This Jupyter Notebook contains all necessary steps in order to set up the dataset Anti-UAV300.

## Dataset Anti-UAV300

The Anti-UAV300 Dataset provided by Nan Jiang et. al(2021)[1] contains 300 video sequences, each containing visible light captured via a standard camera sensor and an uncalibrated infrared counterpart. 

<sub>[1] Nan Jiang, Kuiran Wang, Xiaoke Peng, Xuehui Yu, Qiang Wang, Junliang Xing, Guorong Li, Jian Zhao, Guodong Guo, & Zhenjun Han. (2021). Anti-UAV: A Large Multi-Modal Benchmark for UAV Tracking.<sub>

### Download (6.04GB)

```python
import gdown

# Dataset provided via GitHub: https://github.com/ucas-vg/Anti-UAV
url = f"https://drive.google.com/uc?id=1NPYaop35ocVTYWHOYQQHn8YHsM9jmLGr"
output = "datasets/anti-uav300-rawetst.zip"

gdown.download(url, output, quiet=False)
print(f"Raw dataset download completed.")
```

### Unzip

```python
import zipfile
import os

zip_path = "datasets/anti-uav300-raw.zip"
extract_dir = "datasets/anti-uav300"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

print(f"File extracted to {extract_dir}")
```
